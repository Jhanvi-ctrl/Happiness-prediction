```python
fig=plt.figure(figsize=(6,4))
data= [happy_data['Health (Life Expectancy)'], happy_data['Freedom'], happy_data['Trust (Government Corruption)'], happy_data['Dystopia Residual']]
plt.boxplot(data)
plt.show()
```

```python
temp=happy_data.copy()
new_data=temp.drop(['Trust (Government Corruption)'], axis=1)
new_data.head()
```

```python
#segregating target variable and removing useless column
y=new_data['Happiness Rank']
fin_data=new_data.drop(['Region', 'Country', 'Happiness Rank'], axis=1)
fin_data.columns
```

```python
scale=StandardScaler()
data= scale.fit_transform(fin_data)
#saving the standardscale class file
import pickle
pickle.dump(scale,open('/content/drive/MyDrive/Dataset/Happiness.csv','wb'))
#now you can use index for column and rows
print(data[0][0])
```

```python
fig= plt.figure(figsize=(6,4))
inserting=[data[4],data[5],data[7]]
plt.boxplot(inserting)
plt.show()
```

```python
Xtrain,Xtest,ytrain,ytest= train_test_split(data, y, test_size=0.1)
reg=LinearRegression()
reg.get_params(deep=True)
```

```python
vari=reg.fit(Xtrain, ytrain)
print(reg.score(Xtrain,ytrain))
```

```python
pred=reg.predict(Xtest)
r2_score=reg.score(Xtest,ytest)
print(r2_score)
```

```python
saved_model=pickle.dump(reg,open('/content/drive/MyDrive/Dataset/HappinessModel.pickle','wb'))
```
